```python
# Make HTTP requests for internet reqources.
import urllib.request

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book.
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]

# Get a sample paragraph - I looked for this by hand.
paragraph = ' '.join(book[58:63])

# Show the paragraph.
print(paragraph)
```

```python
# Let's lower-case it.
alice = paragraph.lower()

# All letters and a space.
chars = 'abcdefghijklmnopqrstuvwxyz '

# And strip anything that is not a letter or space.
alice = ''.join([c for c in alice if c in chars])

# Show the paragraph now.
print(alice)
```

```python
# For doing (pseudo-)random things in Python.
import random

# Get the length of alice.
N = len(alice)

# Generate N random characters from chars.
gener = random.choices(chars, k=N)

# Join them together in a string.
gener = ''.join(gener)

# Print.
print(gener)
```

```python
# Get the whole book in one big string.
sbook = ''.join(book[26:]).lower()

# Create the weights - count the occurences of each character in the whole book.
weights = [sbook.count(c) for c in chars]

# Show the weights.
weights
```

```python
# Generate a string using those weights.
wgenr = random.choices(chars, weights=weights, k=N)

# Join them together in a string.
wgenr = ''.join(wgenr)

# Print.
print(wgenr)
```

```python
# Create the 3D weights.
three_weights = {c: {d: {e: sbook.count(c + d + e) for e in chars} for d in chars} for c in chars}

# Show the weights.
three_weights
```

```python
# Loop through our character set.
for i in range(len(chars)):
    # Print the character and how many times it appears in Alice in Wonderland.
    print(f'{chars[i]}: {weights[i]}')
```

```python
# Start with space.
pairs = ' e'

str_len = 1000
# Do the following str-len times.
for i in range(1, str_len):
    # Get the weights where the previous character is the second last character in threesWeight.
    wt = three_weights[pairs[-2]]
    # Now get the weights where the previous character is the last character in threesWeight.
    wt = wt[pairs[-1]]
    # Turn wt into a list, ordered by chars.
    wt = [wt[c] for c in chars]
    # Randomly pick the next character using those weights.
    nextc = random.choices(chars, weights=wt, k=1)[0]
    # Append the character to twos.
    pairs = pairs + nextc

pairs
```

```python
### ENTROPY ###

# Trying out code from Class
import math
import matplotlib.pyplot as plt
import numpy as np

# I(M) = log2(1/p(M)) = -log2 p(M)bits
# with bts it's 1 or 0 hence: I(M) = log2(1/2) = -log2 2 = 1 bit

print("### 50/50 probability using math.log ###")
print(-math.log(0.5,2))

# What if 1 is received with the probability of 1/4?
# I(M) = -log2 1/4 = log2 4 = 2 bits
print("\n### 1/4 probability using math.log ###")
print(-math.log(0.25, 2))

# What if 1 is received with the probability of 3/4?
# I(M) = -log2 3/4 = log2 4/3 = 0.4 bits
print("\n### 3/4 probability using math.log ###")
print(-math.log(0.75, 2))

# 50/50 bit using average information content for all possible messages
print("\n### 50/50 probability using the average for all possible messages ###")
# For a 50/50 bit => -(0.5 x log2 1/2 + 0.5 x log2 1/2) = 1 bit
p = [0.5, 0.5]

ans = -sum([p_i * math.log(p_i, 2) for p_i in p])
print(ans)

# 1/4 bit using average information content for all possible messages
print("\n### 1/4 probability using the average for all possible messages ###")
# For a 50/50 bit => -(0.5 x log2 1/2 + 0.5 x log2 1/2) = 1 bit
p = [0.25, 0.75]

ans = -sum([p_i * math.log(p_i, 2) for p_i in p])
print(ans)

# Plotting the range of possible values without zero (See below)
p_of_1 = np.linspace(0.01, 0.99, 1000)
entropy = - (p_of_1 * np.log2(p_of_1) + (1.0 - p_of_1) * np.log2(1.0 - p_of_1))

# Create the plot
fig, ax = plt.subplots(figsize=(14,6))
ax.plot(p_of_1, entropy, color='#ff791e')

# Label axes
ax.set_xlabel('Probability the bit is 1, not 0')
ax.set_ylabel('Entropy')
```

```python
### Fire Alarms ###

# Code from class
mins_in_year = 365 * 24 * 60

# probability of screeching
p_screech = 10/mins_in_year

# Information content of a screech
-math.log(1.0 - p_screech, 2)

# Entropy of the fire alarm as a source of information
p = [p_screech, 1.0 - p_screech]
entropy = -sum([p_i * math.log(p_i, 2) for p_i in p])
print(f'{entropy} bits')
```

```python
### PIN Codes ###

# Code from class
# four digit PIN has 10,000 possible values.
-math.log(1.0/10000.0, 2)

# Entropy of four digit PIN
print("\n### Entropy of four digit PIN ###")
p = [1.0 / 10000.0 for i in range(10000)]
entropy = -sum([p_i * math.log(p_i, 2) for p_i in p])
print(f'{entropy} bits')

# Entropy of six digit PIN
print("\n### Entropy of six digit PIN ###")
p = [1.0 / 1000000.0 for i in range(1000000)]
entropy = -sum([p_i * math.log(p_i, 2) for p_i in p])
print(f'{entropy} bits')
```

```python
Why is the log of zero undefined?
---------------------------------
Logarithms Explained
--------------------
Log is written as logbase exponent e.g. log2 4 where base = 2 and exponent = 4
What log is saying is what do I raise the base to the power of, to get the exponent.
So, in the example what power do I have to raise 2, to get 4.
We all know 2 to the power of 2 is 4.
Therefore log2 4 = 2

So, let’s look at zero…
??? Logbase 0 ???
Regardless of the base, what do I raise the base to the power of, to get 0.
You can never get zero using powers, any number to the power of 0 is 1, any number to the power of 1 is itself.
So you can never get zero!
That is whay the log of zero is undefined.
```
