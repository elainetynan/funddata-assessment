# Statistical Bias
****
Two of the common calculation sn statistics are mean and standard deviation.

The mean is straight-forward.

## Mean

Suppose you take a sample of the values from a larger population of values.

If you take the mean of the sample, it is a good estimate of the population.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generaye a sample of 1000 values from a normal distribution
x = np.random.normal(10.0, 1.0, 1000) # mean = 10, standard deviation (spread) = 1, n = 1000
x.mean()
```

```python
# 100 samples of 1000
samples = np.random.normal(10.0, 1.0, (1000, 1000))
samples
```

```python
# Mean of first sample
samples[0].mean()
```

```python
# Mean of all samples
sample_means = samples.mean(axis=1)
sample_means
```

```python
plt.hist(sample_means)
```

## Standard Deviation

Standard Deviation is a different story.

It gives a measure of how far the numbers are away frinm the mean in general.

Both below set of numbers have the same mean.

```python
numbers1 = np.array([1,1,1,1,10])
np.mean(numbers1)
```

```python
numbers2 = np.array([2,2,3,3,4])
np.mean(numbers2)
```

```python
# Get the zeroed values (subtract the mean from each of the values)
x_mean = x.mean()
zeroed = x - x_mean
zeroed
```

```python
# mean od zeroed data should be 0 (except for rounding errors)
zeroed.mean()
```

```python
print(f'{zeroed.mean():0.4f}')
```
